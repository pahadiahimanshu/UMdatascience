## Pandas have variety of statistical function

```python
print(early_finisher['assignment1_grade'].mean())
print(late_finisher['assignment1_grade'].mean())
```

- Similar means. But, are they the same? 
- This is where the students' t-test comes in. It allows us to form the alternative hypothesis ("These are different") as well as the null hypothesis ("These are same") and then test that null hypothesis.

- When doing hypothesis testing, we have to choose a significance level as a threshold for how much of a chance we're willing to accept. This significance level is called alpha.
- For eg. alpha = 0.05 or 5%

 ### SciPy's ttest_ind() function
 - **Result**
    - t-statistic and a p-value
    - p-value is more important as it indicates the chance of our Null hypothesis being true

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(early_finisher['assignment1_grade'], late_finisher['assignment1_grade'])
```

```python
# probability is 0.18 which is > than our alpha (0.05). This means we cannot reject null hypothesis.
# Null hypothesis was the two populations are same, and we don't have enough certainty in our evidencve. 
# Because prob > alpha.
```

```python
print(ttest_ind(early_finisher['assignment2_grade'], late_finisher['assignment2_grade']))
print(ttest_ind(early_finisher['assignment3_grade'], late_finisher['assignment3_grade']))
print(ttest_ind(early_finisher['assignment4_grade'], late_finisher['assignment4_grade']))
print(ttest_ind(early_finisher['assignment5_grade'], late_finisher['assignment5_grade']))
print(ttest_ind(early_finisher['assignment6_grade'], late_finisher['assignment6_grade']))
```

## Are these two dataframes same?


```python
# Alpha is 0.1 or 10 %

def test_columns(alpha=0.1):
    # no of columns that differ
    num_diff=0
    for col in df1.columns:
        teststat, pval = ttest_ind(df1[col],df2[col])
        if pval <= alpha:
            print('Col {} is statistically significantly different at alpha={} and pval={}'.format(col,alpha, pval))
            num_diff = num_diff+1
    print('Total number of different was {}, which is {}%'.format(num_diff, float(num_diff)/len(df1.columns)*100))

test_columns()
```
